#  Ecuación maestra

Considera una caminata aleatoria en una dimensión. Hasta ahora, hemos calculado **trayectorias individuales** (realizaciones) del proceso.

Ahora veamos al sistema desde un punto de vista "dual", en términos ya no de las trayectorias individuales de las partículas, sino en caracterizar la *distribución de probabilidad* completa.

Sea $P_t(i)$ la probabilidad de que el caminante se encuentre en el sitio $i$ al tiempo (discreto) $t \in \mathbb{N}$.  Para una $t$ dada, nos interesa el conjunto de valores $\{ P_t(i)$: $i \in \mathbb{Z} \}$,
es decir la *distribución de probabilidad* (o, estrictamente, *función de masa*) al tiempo $t$. Podemos pensar en este objeto como un vector $\mathbf{P}_t$ (con, en principio, ¡un número *infinito* de entradas!)

```julia
Dim = 5; T = 4
Red = zeros(Dim)
Origen = int(ceil(Dim/2))
Red[Origen] = 1.

p = 0.5; q = 1 - p
D = zeros(Dim)
for t in 1:T
    CopiaRed = zeros(Dim)
    CopiaRed[end] = p*Red[end-1]
    CopiaRed[1] = q*Red[2]
    for i in 2:Dim-1
        CopiaRed[i] = q*Red[i-1]+p*Red[i+1]
    end
    CopiaRed[2] += p*Red[1]
    CopiaRed[end-1] += q*Red[end]
    Red = CopiaRed
    show(Red)
    D = hcat(D, Red)
end
transpose(D)
```

```julia
function DistProbCamAleat1D(Dim, T, p)
    Red = zeros(Dim)
    Origen = int(ceil(Dim/2))
    Red[Origen] = 1.

    q = 1 - p
    D = zeros(Dim)
    for t in 1:T
        CopiaRed = zeros(Dim)
        CopiaRed[end] = p*Red[end-1]
        CopiaRed[1] = q*Red[2]
        for i in 2:Dim-1
            CopiaRed[i] = q*Red[i-1]+p*Red[i+1]
        end
        CopiaRed[2] += p*Red[1]
        CopiaRed[end-1] += q*Red[end]
        Red = CopiaRed
        D = hcat(D, Red)
    end
    return transpose(D)
end
```

```julia
D = DistProbCamAleat1D(71, 100, 0.5)
imshow(D)
```

```julia
function Proba_Cam_Aleat_2D_1(Dim, px, py, OrigenX, OrigenY, T)
    Red = zeros(Dim+2, Dim+2)
    Red[OrigenX+1, OrigenY+1] = 1.

    V = Any[]
    qx = 1 - px
    qy = 1 - py
    D = zeros(Dim, Dim)
    D[OrigenX, OrigenY] = 1
    push!(V, D)
    
    for t in 1:T
        CopiaRed = zeros(Dim+2, Dim+2)
        for x in 2:Dim+1
            a = 1; b = 1
            if x-1 == 2
                a = 1/px
                b = 0
            end

            if x+1 == Dim+1
                b = 1/qx
                a = 0
            end
        
            for y in 2:Dim+1
                c = 1; d = 1
                if y-1 == 2
                    c = 1/py
                    d = 0
                end
            
                if y+1 == Dim+1
                    d = 1/qy
                    c = 0
                end
            
                CopiaRed[x, y] = 0.5*(a*px*Red[x-1, y] + b*qx*Red[x+1, y] + c*py*Red[x, y-1] + d*qy*Red[x, y+1])
            end
        end
        Red = CopiaRed
        B = sub(Red, 2:1:Dim+1, 2:1:Dim+1)
        push!(V, B)
    end
    return V
end
```
